Розширемо гіпотезу, чи варто зосередити маркетингові зусилля на залученні жінок до програми учасників. Для цього перевіремо гіпотези 2.1 і 2.2

**2.1 Чи учасники програми дійсно купують більше (вони мають більшу кількість інвойсів), витрачають більше грошей за покупку (більший середній чек) і приносять більше валового доходу, ніж не-учасники програми?**

Учасники програми дійсно купують більше (вони мають більшу кількість інвойсів), витрачають більше грошей за покупку (більший середній чек) і приносять більше валового доходу, ніж не-учасники програми. Ці висновки можуть допомогти в розумінні поведінки покупців та оптимізації маркетингових та продажних стратегій.

```python
# Розділення даних на дві групи: Members і Normals
members_data = df[df['Customer type'] == 'Member']
normals_data = df[df['Customer type'] == 'Normal']

# Розрахунок статистик для Members
invoices_members = len(members_data)
average_bill_members = members_data['Total'].mean()
average_gross_income_members = members_data['gross income'].mean()

# Розрахунок статистик для Normals
invoices_normals = len(normals_data)
average_bill_normals = normals_data['Total'].mean()
average_gross_income_normals = normals_data['gross income'].mean()

# Виведення результатів
print("Статистика для Members:")
print(f"Кількість інвойсів: {invoices_members}")
print(f"Середній чек: {average_bill_members}")
print(f"Середній валовий дохід: {average_gross_income_members}")

print("Статистика для Normals:")
print(f"Кількість інвойсів: {invoices_normals}")
print(f"Середній чек: {average_bill_normals}")
print(f"Середній валовий дохід: {average_gross_income_normals}")
```

**3. Як відгуки клієнтів пов'язані з кількістю продажів у кожному місті?**

Гіпотеза: Вищі середні оцінки відгуків в одному місті можуть корелювати з вищими продажами в тому місті.


Дані показують, що сам кращий рейтинг впливає більшу суму продажів, але не так сильно впливає на кількість продажів. Naypyitaw має найвищий рейтинг та найвищий середній чек, ймовірно  його хороша репутація приваблює преміальний сегмент клієнтів.

```python
city_ratings_sales = df.groupby('City').agg({'Rating': 'mean', 'Invoice ID': 'count'})
city_ratings_sales.round(2).sort_values('Invoice ID', ascending=False)
```

```python
city_ratings_sales = df.groupby('City').agg({'Rating': 'mean', 'Total': 'sum'})
city_ratings_sales.round(2).sort_values('Total', ascending=False)
```

```python
df.groupby('City').agg({'Total': 'mean'}).round(2).sort_values('Total', ascending=False)
```
