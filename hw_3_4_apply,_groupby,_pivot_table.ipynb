–£ –ø–æ–Ω–µ–¥—ñ–ª–æ–∫ –±—ñ–ª—å—à–µ –∫—É–ø—É—é—Ç—å Health and beauty.

–°—É–±–æ—Ç–∞ - Fashion accessories.

–ù–µ–¥—ñ–ª—è - Home and lifestyle.





```python
df_copy = df
df_copy['Date'] = pd.to_datetime(df_copy['Date'])
df_copy['DayOfWeek'] = df_copy['Date'].dt.day_name()
sales_by_day_week = df_copy.groupby(['DayOfWeek']).agg({'Total': 'sum'})
sales_by_day_week.round(2).sort_values('Total', ascending=False)
```

```python
df_copy['DayOfWeek1'] = df_copy['Date'].dt.dayofweek + 1
category_sales_by_day_week = df_copy.groupby(['DayOfWeek1', 'Product line']).agg({'Total': 'sum'})
category_sales_by_day_week.round(2).sort_values(['DayOfWeek1','Total'], ascending=[True, False])
```

**2.1 –ß–∏ —É—á–∞—Å–Ω–∏–∫–∏ –ø—Ä–æ–≥—Ä–∞–º–∏ –¥—ñ–π—Å–Ω–æ –∫—É–ø—É—é—Ç—å –±—ñ–ª—å—à–µ (–≤–æ–Ω–∏ –º–∞—é—Ç—å –±—ñ–ª—å—à—É –∫—ñ–ª—å–∫—ñ—Å—Ç—å —ñ–Ω–≤–æ–π—Å—ñ–≤), –≤–∏—Ç—Ä–∞—á–∞—é—Ç—å –±—ñ–ª—å—à–µ –≥—Ä–æ—à–µ–π –∑–∞ –ø–æ–∫—É–ø–∫—É (–±—ñ–ª—å—à–∏–π —Å–µ—Ä–µ–¥–Ω—ñ–π —á–µ–∫) —ñ –ø—Ä–∏–Ω–æ—Å—è—Ç—å –±—ñ–ª—å—à–µ –≤–∞–ª–æ–≤–æ–≥–æ –¥–æ—Ö–æ–¥—É, –Ω—ñ–∂ –Ω–µ-—É—á–∞—Å–Ω–∏–∫–∏ –ø—Ä–æ–≥—Ä–∞–º–∏?**

–£—á–∞—Å–Ω–∏–∫–∏ –ø—Ä–æ–≥—Ä–∞–º–∏ –¥—ñ–π—Å–Ω–æ –∫—É–ø—É—é—Ç—å –±—ñ–ª—å—à–µ (–≤–æ–Ω–∏ –º–∞—é—Ç—å –±—ñ–ª—å—à—É –∫—ñ–ª—å–∫—ñ—Å—Ç—å —ñ–Ω–≤–æ–π—Å—ñ–≤), –≤–∏—Ç—Ä–∞—á–∞—é—Ç—å –±—ñ–ª—å—à–µ –≥—Ä–æ—à–µ–π –∑–∞ –ø–æ–∫—É–ø–∫—É (–±—ñ–ª—å—à–∏–π —Å–µ—Ä–µ–¥–Ω—ñ–π —á–µ–∫) —ñ –ø—Ä–∏–Ω–æ—Å—è—Ç—å –±—ñ–ª—å—à–µ –≤–∞–ª–æ–≤–æ–≥–æ –¥–æ—Ö–æ–¥—É, –Ω—ñ–∂ –Ω–µ-—É—á–∞—Å–Ω–∏–∫–∏ –ø—Ä–æ–≥—Ä–∞–º–∏. –¶—ñ –≤–∏—Å–Ω–æ–≤–∫–∏ –º–æ–∂—É—Ç—å –¥–æ–ø–æ–º–æ–≥—Ç–∏ –≤ —Ä–æ–∑—É–º—ñ–Ω–Ω—ñ –ø–æ–≤–µ–¥—ñ–Ω–∫–∏ –ø–æ–∫—É–ø—Ü—ñ–≤ —Ç–∞ –æ–ø—Ç–∏–º—ñ–∑–∞—Ü—ñ—ó –º–∞—Ä–∫–µ—Ç–∏–Ω–≥–æ–≤–∏—Ö —Ç–∞ –ø—Ä–æ–¥–∞–∂–Ω–∏—Ö —Å—Ç—Ä–∞—Ç–µ–≥—ñ–π.

```python
# –†–æ–∑–¥—ñ–ª–µ–Ω–Ω—è –¥–∞–Ω–∏—Ö –Ω–∞ –¥–≤—ñ –≥—Ä—É–ø–∏: Members —ñ Normals
members_data = df[df['Customer type'] == 'Member']
normals_data = df[df['Customer type'] == 'Normal']

# –†–æ–∑—Ä–∞—Ö—É–Ω–æ–∫ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫ –¥–ª—è Members
invoices_members = len(members_data)
average_bill_members = members_data['Total'].mean()
average_gross_income_members = members_data['gross income'].mean()

# –†–æ–∑—Ä–∞—Ö—É–Ω–æ–∫ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫ –¥–ª—è Normals
invoices_normals = len(normals_data)
average_bill_normals = normals_data['Total'].mean()
average_gross_income_normals = normals_data['gross income'].mean()

# –í–∏–≤–µ–¥–µ–Ω–Ω—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ñ–≤
print("–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –¥–ª—è Members:")
print(f"–ö—ñ–ª—å–∫—ñ—Å—Ç—å —ñ–Ω–≤–æ–π—Å—ñ–≤: {invoices_members}")
print(f"–°–µ—Ä–µ–¥–Ω—ñ–π —á–µ–∫: {average_bill_members}")
print(f"–°–µ—Ä–µ–¥–Ω—ñ–π –≤–∞–ª–æ–≤–∏–π –¥–æ—Ö—ñ–¥: {average_gross_income_members}")

print("–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –¥–ª—è Normals:")
print(f"–ö—ñ–ª—å–∫—ñ—Å—Ç—å —ñ–Ω–≤–æ–π—Å—ñ–≤: {invoices_normals}")
print(f"–°–µ—Ä–µ–¥–Ω—ñ–π —á–µ–∫: {average_bill_normals}")
print(f"–°–µ—Ä–µ–¥–Ω—ñ–π –≤–∞–ª–æ–≤–∏–π –¥–æ—Ö—ñ–¥: {average_gross_income_normals}")
```
