1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
# задаємо функцію
def f(x):
  return 10 * x - 5
# похідна константа
f_derivative = 10

# діапазон значень від 0 до 10
start, end = 0, 10

# створюємо масив і заповнюємо значеннями для функції
x_vals = np.linspace(start, end)
# створюємо масив і заповнюємо значеннями функції
y_vals = f(x_vals)
# створюємо масив і заповнюємо значеннями похідної, це завжди 10
y_derivative_vals = np.full_like(x_vals, f_derivative)

# будуємо графік функції
plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='f(x) = 10x - 5')
# додаємо на графік лінію похідної
plt.plot(x_vals, y_derivative_vals, label="f'(x) = 10", linestyle='--')
plt.title('Залежність оцінки від кількості годин, витрачених на навчання')
plt.xlabel('Години навчання (x)')
plt.ylabel('Оцінка та її зміна з часом підготовки')
plt.legend()
plt.grid(True)
```

###Відповідь:
Оскільки похідна константа, то кожна година навчання збільшує оцінку на 10 одиниць. Якщо учасник не готується взагалі, то стартова оцінка складає -5, але відʼємні оцінки зазвичай не ставлять.
Згідно з графіком, для "трієчки"/60 одиниць оцінки треба витратити не менше 6.5 годин на навчання.


2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
# задаємо змінну та функції
x = sp.symbols('x')
sin_x = sp.sin(x)
ln_x = sp.log(x)
# обчислюємо похідні функцій
sin_x_derivative = sp.diff(sin_x, x)
ln_x_derivative = sp.diff(ln_x, x)
display(sin_x_derivative, ln_x_derivative)
```

```python
f1_func = sp.lambdify(x, sin_x, 'numpy')
f2_func = sp.lambdify(x, ln_x, 'numpy')
# створюємо масиви значень похідної для кожної функції
f1_derivative_func = sp.lambdify(x, sin_x_derivative, 'numpy')
f2_derivative_func = sp.lambdify(x, ln_x_derivative, 'numpy')

# створюємо масиви значень для 𝑓(𝑥)=sin(𝑥)
x_vals = np.linspace(-2*np.pi, 2*np.pi, 400)
# створюємо масиви значень функції 𝑓(𝑥)=sin(𝑥)
y_f1 = f1_func(x_vals)
# створюємо масиви значень похідної 𝑓(𝑥)=sin(𝑥)
y_f1_derivative = f1_derivative_func(x_vals)

# створюємо масиви значень для 𝑓(𝑥)=ln(𝑥)
x_vals_positive = np.linspace(0.1, 4, 400)
# створюємо масиви значень функції 𝑓(𝑥)=ln(𝑥)
y_f2 = f2_func(x_vals_positive)
# створюємо масиви значень похідної 𝑓(𝑥)=ln(𝑥)
y_f2_derivative = f2_derivative_func(x_vals_positive)
```

```python
plt.figure(figsize=(12, 6))
# графік для 𝑓(𝑥)=sin(𝑥)
plt.subplot(1, 2, 1)
plt.plot(x_vals, y_f1, label='sin(x)')
plt.plot(x_vals, y_f1_derivative, label="sin'(x) = cos(x)")
plt.title("Функція sin(x) та її похідна")
plt.xlabel('x')
plt.ylabel('f(x), f\'(x)')
plt.legend()
plt.grid(True)
# графік для 𝑓(𝑥)=ln(𝑥)
plt.subplot(1, 2, 2)
plt.plot(x_vals_positive, y_f2, label='ln(x)')
plt.plot(x_vals_positive, y_f2_derivative, label="ln'(x) = 1/x")
plt.title("Функція ln(x) та її похідна")
plt.xlabel('x')
plt.ylabel('f(x), f\'(x)')
plt.legend()
plt.grid(True)
plt.tight_layout()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')
f = x**2 - 5*x + 1
f_derivative = sp.diff(f, x)

f_func = sp.lambdify(x, f, 'numpy')
f_derivative_func = sp.lambdify(x, f_derivative, 'numpy')

start, end = 0, 100
x_vals = np.linspace(start,end)
y_f = f_func(x_vals)
y_f_derivative = f_derivative_func(x_vals)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_f, label='f(x) = x^2 - 5x + 1')
plt.plot(x_vals, y_f_derivative, label="f'(x) = 2x - 5", linestyle='--')
plt.title('Загальні витрати та їх зміна зі збільшенням виробництва')
plt.xlabel('Кількість вироблених одиниць (x)')
plt.ylabel('Вартість (f(x)) та Зміна вартості (f\'(x))')
plt.legend()
plt.grid(True)
```
